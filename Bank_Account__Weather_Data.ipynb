##  Weather Data Analyzer

```python
# Weather Data Analyzer
region = "Kerala"
temperatures = [32, 34, 31, 33, 35]
humidities = [65, 70, 68, 72, 66]
print("===== WEATHER DATA ANALYZER =====")
print("Region:", region)
print("Temperature Data:", temperatures)
print("Humidity Data:", humidities)
```

```python
average_temperatures = sum(temperatures) / len(temperatures)
average_humidity = sum(humidities) / len(humidities)
def weather_condition(avg_temp, avg_humidity):
    if avg_temp > 30 and avg_humidity < 70:
        return "Sunny"
    elif avg_humidity >= 70:
        return "Rainy"
    else:
        return "Cloudy"

condition = weather_condition(average_temperatures, average_humidity)

print(f"Average Temperature: {avg_temp:}°C")
print(f"Average Humidity: {avg_hum:}%")
print("Weather Condition:", condition)
```
