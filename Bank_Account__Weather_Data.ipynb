```python
total_amount = 10000000
balance = 75000.0

print("\n===== BANK ACCOUNT MANAGER =====")
print("1. Deposit")
print("2. Withdraw")
print("3. Balance Enquiry")

choice = input("Enter your choice (1, 2, or 3): ")

if choice == "1":
    user_deposit = float(input("Enter the deposit amount (in Rupees): "))

    if user_deposit > 0:
        total_amount += user_deposit
        balance += user_deposit

        print("\n--- Deposit Successful ---")
        print(f"Amount Deposited: Rs. {user_deposit:}")
        print(f"Your Updated Account Balance: Rs. {balance:}")
        print(f"Total Branch Deposits: Rs. {total_amount:}")
    else:
        print("Please enter a valid amount.")

elif choice == "2":
    user_withdrawal = float(input("Enter the withdrawal amount (in Rupees): "))

    if user_withdrawal > 0:
        if user_withdrawal <= balance:
            balance -= user_withdrawal
            total_amount -= user_withdrawal

            print("\n--- Withdrawal Successful ---")
            print(f"Amount Withdrawn: Rs. {user_withdrawal:}")
            print(f"Updated Balance: Rs. {balance:}")
            print(f"Total Branch Deposits: Rs. {total_amount:}")
        else:
            print("Insufficient balance.")
    else:
        print("Please enter a valid amount.")

elif choice == "3":
    print("\n--- Balance Enquiry ---")
    print(f"Current Account Balance: Rs. {balance:}")

else:
    print("Invalid choice.")



```

##  Weather Data Analyzer

```python
# Weather Data Analyzer
region = "Trivandrum"
temperatures = [32, 34, 31, 33, 35]
humidities = [65, 70, 68, 72, 66]
```

```python
def average_temperature(temp_list):
    return sum(temp_list) / len(temp_list)

# Function to calculate average humidity
def average_humidity(humidity_list):
    return sum(humidity_list) / len(humidity_list)

# Function to determine weather condition
def weather_condition(avg_temp, avg_humidity):
    if avg_temp > 30 and avg_humidity < 70:
        return "Sunny"
    elif avg_humidity >= 70:
        return "Rainy"
    else:
        return "Cloudy"

# Sample Data
region = "Kerala"
temperatures = [32, 31, 30, 33, 29]
humidities = [65, 68, 72, 70, 66]

# Main Program
avg_temp = average_temperature(temperatures)
avg_hum = average_humidity(humidities)
condition = weather_condition(avg_temp, avg_hum)

# Display Results
print("===== WEATHER DATA ANALYZER =====")
print("Region:", region)
print("Temperature Data:", temperatures)
print("Humidity Data:", humidities)

print(f"\nAverage Temperature: {avg_temp:.2f}°C")
print(f"Average Humidity: {avg_hum:.2f}%")
print("Weather Condition:", condition)
```
